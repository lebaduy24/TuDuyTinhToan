### Bài 1: Tìm số lớn nhất của 1 list

```python
lst = list(map(int,input().split()))
max_value = lst[0]
for x in lst:
    if x > max_value:
        max_value = x
print (max_value)
```

### Bài 7: Kiểm tra một list số nguyên có phải là dãy tăng, dãy giảm hay vô danh
1 3 5 6 7 8 9: Tang
9 8 7 6 5 4 0: Giam
1 2 3 7 1 0 2: Vo danh
1 2 2 4 5 7: Vo danh

```python
lst = list(map(int, input().split()))

is_tang = True
is_giam = True

for i in range(len(lst) - 1):
    if lst[i] >= lst[i + 1]:
        is_tang = False

    if lst[i] <= lst[i + 1]:
        is_giam = False

if is_tang:
    print("Tang")
elif is_giam:
    print("Giam")
else:
    print("Vo danh")
```

### Bài 8: Kiểm tra một list số nguyên có phải là ziczac khong
1 3 2 5 2 8 7 9 2 100: ziczac
34 2 56 7 90 20 30 7 120: ziczac

```python

lst = list(map(int, input("Nhập dãy số: ").split()))

ziczac = True

if len(lst) < 3:
   
    if len(lst) == 2 and lst[0] == lst[1]:
        ziczac = False
else:
    for i in range(len(lst) - 2):
        a = lst[i]
        b = lst[i+1]
        c = lst[i+2]
        
    
        la_dinh = (a < b and b > c)
        la_day = (a > b and b < c)
        
    
        if not (la_dinh or la_day):
            ziczac = False
            break 

if ziczac:
    print("Ziczac")
else:
    print("Không phải Ziczac")
```

### Bài 9: Tìm và in ra tất cả các vị trí của số nguyên k trong một danh sách số nguyên cho trước. Nếu không có k trong ds thì in ra "-1".
VD: k = 3, [1,3,4,4,3,3,4,3,6,1] => k xuất hiện tại các vị trí (trả về 1 list) [1, 4, 5, 7]
k = 10, [1,3,4,4,3,3,4,3,6,1] => k không xuất hiện trong ds => trả về "-1"

```python
danh_sach_so = [1, 5, 3, 5, 7, 8, 5, 2]
so_can_tim = 5

cac_vi_tri = []

for chi_so in range(len(danh_sach_so)):
    if danh_sach_so[chi_so] == so_can_tim:
        cac_vi_tri.append(chi_so)

if len(cac_vi_tri) > 0:
    for vi_tri in cac_vi_tri:
        print(vi_tri)
else:
    print("-1")
```
