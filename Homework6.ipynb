Nhập vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma ̣ trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải

```python
n,m=map(int,input().split())
a=[list(map(int,input().split())) for _ in range(n)]
for n in a:
    print(*n)
```

Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về -1

```python
a=list(map(int,input().split()))
k=int(input())
vt=-1
for i in range(0,len(a)):
    if k==a[i]:
        vt=i
        break
print(vt)
```

Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện

```python
n=input()
a=n.split()
k={}
for i,j in enumerate(a):
    if j not in k:
        k[j]=[]
    k[j].append(i)
print(k)
```

Viết hàm xóa đi các số <x có trong mảng

```python
x=int(input())
a=list(map(int,input().split()))
def xoa():
    global a
    m=[]
    for i in a:
        if i>=x:
            m.append(i)
    print(m)
xoa()
```

Viết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 2 3 2 4 9

```python
a=list(map(int,input().split()))
b=[]
d=0
for i in a:
    if i !=0:
        b.append(i)
    else:d+=1
for i in range(0,d):
    b.insert(0,0)
print(b)
```

Cho dãy chỉ gồm N giá trị từ 0 - 9 . Kiểm tra xem giá trị nào không có trong mảng

```python
a=list(map(int,input().split()))
m=[0]*10
for i in a:
    m[i]+=1
for i in range(len(m)):
    if m[i]==0:print(i,end=" ")
```

Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2 phần tử đầu và cuối

```python
a=list(map(int,input().split()))
s=0
for i in range(1,len(a)-1):
    if a[i]>a[i-1] and a[i]>a[i+1]:s+=a[i]
print(s)
```

Cho một dãy số nguyên, Tìm và in ra vị trí các số < x (x là giá trị cho trước). nếu không có in ra -1

```python
x=int(input())
a=list(map(int,input().split()))
m=[]
for i in range(len(a)):
    if a[i]<x:
        m.append(i+1)
if len(m)==0:print("-1")
else:print(m)
```

Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm thực sự, hay vô danh (không tăng, không giảm)

```python
a=list(map(int,input().split()))
k1=0
k2=0
for i in range(0,len(a)-1):
    if a[i]<=a[i+1]:
        k1=1
        break
for i in range(0,len(a)-1):
    if a[i]>=a[i+1]:
        k2=1
        break
if k1==0:print("giam")
if k2==0:print("tang")
if k1!=0 and k2!=0:print("vodanh")
```
