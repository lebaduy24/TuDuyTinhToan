Cho một mảng số nguyên không âm. Gọi mảng đó là "mảng đẹp" nếu 
mỗi phần tử x trong mảng xuất hiện đúng x lần. 
Kiểm tra xem mảng có phải là mảng đẹp hay không. Nếu không hãy xóa 
đi y số (y là nhỏ nhất có thể) để mảng còn lại thành mảng đẹp. 

```python
a = list(map(int, input().split()))
d = {}
for x in a:
    d[x] = d.get(x, 0) + 1
y = 0
for x, count in d.items():
    if x == 0:
        y += count
    elif count >= x:
        y += (count - x)
    else:
        y += count
if y!=0:print("No",y)
else :print("Yes")
```

Tìm phần tử xuất hiện nhiều hơn n/2 lần trong mảng (nếu có) với độ 
phức tạp thời gian là O(n), độ phức tạp bộ nhớ là O(1), - Boyer–Moore. 
Nếu không có phần tử nào thỏa mãn thì không in gì 

```python
a = list(map(int, input().split()))
n = len(a)
d={}
for i in a:
    d[i]=d.get(i,0)+1
for i,j in d.items():
    if j>(n/2):
        print(i)
```

Sắp xếp có điều kiện: Số chẵn đứng trước, số lẻ đứng sau, chẵn tăng 
dần, lẻ giảm dần; ;  

```python
a=list(map(int,input().split()))
b=[]
c=[]
for i in a:
    if i%2==0:
        b.append(i)
    else:c.append(i)
for i in range(len(b)):
    for j in range(0, len(b) - i - 1):
        if b[j] > b[j+1]:
            b[j], b[j+1] = b[j+1], b[j]
for i in range(len(c)):
    for j in range(0, len(c) - i - 1):
        if c[j] < c[j+1]:
            c[j], c[j+1] = c[j+1], c[j]
print(b+c)
```

Merge 2 mảng đã sắp xếp (có thể tăng dần, có thể giảm dần) mà không dùng bất cứ thuật toán sắp xếp nào thành một mảng cũng có cùng thứ tự sắp xếp như 2 mảng ban đầu

```python
a = list(map(int, input().split()))
b = list(map(int, input().split()))
c = True
if len(a) > 1:
    c = a[0] <= a[-1]
elif len(b) > 1:
    c = b[0] <= b[-1]
d, e = len(a), len(b)
f, g = 0, 0
h = []
while f < d and g < e:
    if c:
        i = a[f] <= b[g]
    else:
        i = a[f] >= b[g]
    if i:
        h.append(a[f])
        f += 1
    else:
        h.append(b[g])
        g += 1
h.extend(a[f:])
h.extend(b[g:])
print(*h)
```

Đếm số lượng tên trong xâu kí tự. Tên được định nghĩa là xâu có độ dài >=2 và có chữ cái đầu viết hoa, các chữ cái khác viết thường, chỉ có kí tự cái.

```python
a = input().split()
b = 0
for c in a:
    if len(c) < 2:
        continue
    d = True
    if not ('A' <= c[0] <= 'Z'):
        d = False
    for e in range(1, len(c)):
        if not ('a' <= c[e] <= 'z'):
            d = False
            break
    if d:
        b += 1
print(b)
```

Tìm số có đúng k chữ số lớn nhất có thể trong một số có n chữ số (coi đầu vào là một số rất lớn - str). Không được hoán đổi vị trí các chữ số

```python
a = input().strip()
b = int(input())
c = len(a) - b
d = []
for e in a:
    while c > 0 and d and d[-1] < e:
        d.pop()
        c -= 1
    d.append(e)
print("".join(d[:b]))
```

Cho một dãy ngoặc ()(){}{}[][]..., kiểm tra xem dãy ngoặc đã cho có hợp lệ hay không

```python
a = input()
b = []
c = {')': '(', '}': '{', ']': '['}
d = True
for e in a:
    if e in c.values():
        b.append(e)
    elif e in c:
        if not b or b.pop() != c[e]:
            d = False
            break
    else:
        d = False
        break
if d and not b:
    print("Hợp lệ")
else:
    print("Không hợp lệ")
```

Bài toán Tìm dãy các hợp số dài nhất liên tiếp không chứa số nguyên nào <= N (N là số nguyên dương cho trước). Nếu có nhiều dãy cùng độ dài thì lấy dãy nhỏ nhất (gặp đầu tiên từ trái qua)

```python
n=int(input())
def nto(n):
    if n<2:return False
    if n in [2,3,5,7]:return True
    if n%2==0 or n%3==0 or n%5==0 or n%7==0:return False
    i=5
    while i*i<=n:
        i+=6
        if n%i==0 or n%(i+2)==0:
            return False
    return True
a=[0]*(n+1)
k=0
for i in range(2,n+1):
    if not nto(i):
        a[i]=a[i-1]+1
        k=max(k,a[i])
vt=0
for i in range(2,n):
    if k==a[i]:
        vt=i
        break
for i in range(vt-k+1,vt+1):
    print(i,end=" ")
```

Tìm Dãy con tăng dần liên tục dài nhất của một danh sách nhập vào từ bàn phím. Lấy dãy đầu tiên nếu có nhiều dãy có số phần tử bằng nhau

Tìm Dãy con tăng dần KHÔNG liên tục dài nhất của một danh sách nhập vào từ bàn phím

```python
a = list(map(int, input().split()))
k = 1 
vt = 0
L = [1] * (len(a))
for i in range(1, len(a)):
    for j in range(0, i):
        if a[i] > a[j] and L[i] < L[j] + 1:
            L[i] = L[j] + 1
            if k < L[i]:
                k = L[i]
                vt = i
res = []
tam = k
tg =99999999
for i in range(vt, -1, -1):
    if L[i] == tam and a[i] < tg:
        res.append(a[i])
        tam -= 1
        tg = a[i]
print(*res[::-1])
```
