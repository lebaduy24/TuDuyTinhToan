 
Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số 
nguyên x 

```python
def kiem_tra_armstrong(x):
    so_chu_so = len(str(x))
    tong = 0
    tam = x
    while tam > 0:
        tong = tong + (tam % 10) ** so_chu_so
        tam = tam // 10
    return tong == x
```

Viết hàm kiêểm tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số 
nguyên x

```python
def kiem_tra_chinh_phuong(x):
    if x < 0: return False
    can = int(x**0.5)
    return can * can == x
```

W5A18 VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên. 


```python
def kiem_tra_doi_xung(x):
    if x < 0: return False
    so_dao = 0
    tam = x
    while tam > 0:
        so_dao = so_dao * 10 + tam % 10
        tam = tam // 10
    return so_dao == x
```

W5A19 VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên. 


```python
def kiem_tra_ky_dieu(x):
    while x > 9:
        tong = 0
        for chu in str(x):
            tong = tong + int(chu)
        x = tong
    return x == 1
```

W5A20 Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'. 


```python
def detect_number_type(s):
    try:
        int(s)
        return 'int'
    except:
        try:
            float(s)
            return 'float'
        except:
            return 'neither'
```

W5A21 Viết hàm is_one_product_of_others(a, b, c) trả về True nếu một trong ba số bằng tích của hai số còn lại 


```python
def is_one_product_of_others(a, b, c):
    if a*b == c or a*c == b or b*c == a:
        return True
    return False
```

W5A23 Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình a*x^2 + b*x + c = 0. 


```python
def quadratic_roots_count(a, b, c):
    delta = b*b - 4*a*c
    if delta > 0: return 2
    if delta == 0: return 1
    return 0
```

W5A24 Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không. 


```python
def are_collinear(x1, y1, x2, y2, x3, y3):
    return (y2 - y1) * (x3 - x2) == (y3 - y2) * (x2 - x1)
```

Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay 
không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3), 
(0,3)], p = (2,1)

```python
def point_in_quadrilateral(quad, p):
    x, y = p
    n = len(quad)
    trong = False
    for i in range(n):
        x1, y1 = quad[i]
        x2, y2 = quad[(i + 1) % n]
        if min(y1, y2) < y <= max(y1, y2):
            giao = x1 + (y - y1) * (x2 - x1) / (y2 - y1)
            if x < giao:
                trong = not trong
    return trong
```

W5A27 Viết hàm trả về số lượng chữ số khác 0 trong n số nguyên nhập vào từ bàn phím (một list)

W5A28 Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3). 


```python
import math
def tinh_goc(x1, y1, x2, y2, x3, y3):
    ba = [x1 - x2, y1 - y2]
    bc = [x3 - x2, y3 - y2]
    tich_vo_huong = ba[0] * bc[0] + ba[1] * bc[1]
    do_dai_ba = math.sqrt(ba[0]**2 + ba[1]**2)
    do_dai_bc = math.sqrt(bc[0]**2 + bc[1]**2)
    cos_b = tich_vo_huong / (do_dai_ba * do_dai_bc)
    return math.degrees(math.acos(cos_b))
```

W5A29 Viết hàm trả về giá trị lớn nhất và nhỏ nhất của ba số a,b,c dưới dạng list [max, min].

W5A30 Viết hàm tính hệ số nhị thức C(n, k). 


```python
def tinh_c_n_k(n, k):
    tu = 1
    for i in range(1, n + 1): tu *= i
    mau1 = 1
    for i in range(1, k + 1): mau1 *= i
    mau2 = 1
    for i in range(1, n - k + 1): mau2 *= i
    return tu // (mau1 * mau2)
```

Viết hàm smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính tổng tiền taxi theo quy tắc giá mở cửa, 
giá km, phí chờ, phụ phí ban đêm và giảm giá.

```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    gia_mo_cua = 12000 if car_type == 4 else 15000
    phi_km = km * 11000
    phi_cho = wait_minutes * 500
    tong = gia_mo_cua + phi_km + phi_cho
    if 22 <= start_hour or start_hour <= 5:
        tong = tong * 1.1
    if km > 30:
        tong = tong * 0.9
    return int(tong)
```


Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau 
Nếu n ≤ 0 trả về 0. 

```python
def sum_1_to_n(n):
    if n <= 0: return 0
    tong = 0
    for x in range(1, n + 1):
        s_x = str(x)
        if len(set(s_x)) == len(s_x):
            tong = tong + x
    return tong
```
