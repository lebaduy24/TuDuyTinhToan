VIết chương trình nhập vào một số tới khi được một số nguyên dương thì thôi. 
Kiểm tra số này có phải là số nguyên tố hay không?

```python
import math
while True:
    x = int(input("Nhập x: "))
    if x > 0:
        break
if x < 2:
    print(f"{x} không phải là số nguyên tố")
else:
    for i in range(2, int(math.sqrt(x)) + 1):
        if x % i == 0:
            print(f"{x} không phải là số nguyên tố")
            break
    else:
        print(f"{x} là số nguyên tố")
```

Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không

```python
n = int(input())
if n > 0 and (n & (n - 1) == 0):
    print(f"{n} là số dạng 2^k")
else:
    print(f"{n} không phải số dạng 2^k")
```

Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
n = int(input())
so_chan = 0
so_le = 0
while n > 0:
    chu_so = n % 10
    if chu_so % 2 == 0:
        so_chan = so_chan + 1
    else:
        so_le = so_le + 1
    n = n // 10
print(so_chan)
print(so_le)
```

Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập 

```python
so_nhap = int(input())
if so_nhap != -1:
    lon_nhat = so_nhap
    nho_nhat = so_nhap
    while True:
        so_nhap = int(input())
        if so_nhap == -1:
            break
        if so_nhap > lon_nhat:
            lon_nhat = so_nhap
        if so_nhap < nho_nhat:
            nho_nhat = so_nhap
    print(lon_nhat)
    print(nho_nhat)
```

[Fibonaci1] 
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ 
bằng tổng hai số trước đó 
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt 
quá A 

```python
a = int(input())
f1 = 1
f2 = 1
f_tiep_theo = 1
while f1 + f2 <= a:
    f_tiep_theo = f1 + f2
    f1 = f2
    f2 = f_tiep_theo
print(f_tiep_theo)
```

Cho một chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa, hãy đổi ký tự hoa thành 
thường và thường thành hoa. 

```python
chuoi = input()
ket_qua = ""
for ky_tu in chuoi:
    if "a" <= ky_tu <= "z":
        ket_qua = ket_qua + ky_tu.upper()
    elif "A" <= ky_tu <= "Z":
        ket_qua = ket_qua + ky_tu.lower()
    else:
        ket_qua = ket_qua + ky_tu
print(ket_qua)
```

Nhập vào một chuỗi có dạng 3 số nguyên, mỗi số nguyên cách nhau một dấu phẩy, hãy tính 
tổng 3 số nguyên đó 
VD: 
Nhập: 3, 12, 15 
Tổng: 30

```python
chuoi = input()
tong = 0
so_tam = ""
for ky_tu in chuoi:
    if ky_tu != "," and ky_tu != " ":
        so_tam = so_tam + ky_tu
    else:
        if so_tam != "":
            tong = tong + int(so_tam)
            so_tam = ""
if so_tam != "":
    tong = tong + int(so_tam)
print(tong)
```

Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in 
thường, bao nhiêu ký tự số

```python
chuoi = input()
hoa = 0
thuong = 0
so = 0
for ky_tu in chuoi:
    if "A" <= ky_tu <= "Z":
        hoa = hoa + 1
    elif "a" <= ky_tu <= "z":
        thuong = thuong + 1
    elif "0" <= ky_tu <= "9":
        so = so + 1
print(hoa)
print(thuong)
print(so)
```
