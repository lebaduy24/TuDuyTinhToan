Yêu cầu. Cước phí Taxi Thông Minh. 
Viết một hàm duy nhất smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính và trả về total (tổng tiền phải trả, 
kiểu số nguyên, đơn vị VND). 
 
Các quy tắc cần cài đặt: 
1. Giá mở cửa (0–0.8 km): - Xe 4 chỗ: 11,000 VND - Xe 7 chỗ: 12,000 VND 
2. Giá theo km sau 0.8 km: - Xe 4 chỗ: từ 0.8→20 km nh 13,000 VND/km; trên 20 km tính 10,500 VND/km - Xe 7 chỗ: từ 0.8→20 km nh 14,000 VND/km; trên 20 km tính 12,000 VND/km 
3. Phí chờ: 5 phút đầu miễn phí; sau đó, mỗi 5 phút (làm tròn lên theo từng block 5 phút) tính 2,000 VND. 
4. Phụ thu ban đêm: nếu start_hour thuộc [22,23,0,1,2,3,4,5], cộng +20% trên tiền tạm tính (mở cửa + quãng đường + 
chờ). 
5. Điều kiện đặc biệt: - Nếu km > 30 và không phải ban đêm, áp dụng giảm 10% trên tổng tiền. - Nếu tổng tiền cuối cùng < 50,000 nhưng km > 5, đặt tổng tiền bằng 50,000 VND (mức tối thiểu).

```python
import math

def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    gia = {4: [11000, 13000, 10500], 7: [12000, 14000, 12000]}[car_type]
    
    tien_xe = gia[0]
    if km > 0.8:
        km_duoi_20 = min(km, 20) - 0.8
        km_tren_20 = max(0, km - 20)
        tien_xe += km_duoi_20 * gia[1] + km_tren_20 * gia[2]

    tien_cho = math.ceil(max(0, wait_minutes - 5) / 5) * 2000
    tong = tien_xe + tien_cho
    
    la_dem = start_hour in [22, 23, 0, 1, 2, 3, 4, 5]
    
    if la_dem:
        tong *= 1.2
    elif km > 30:
        tong *= 0.9
        
    if tong < 50000 and km > 5:
        tong = 50000
        
    return int(tong)

print(smart_taxi_fare(4, 10, 10, 0))
print(smart_taxi_fare(4, 23, 10, 10))
print(smart_taxi_fare(7, 12, 35, 0))
```

Yêu cầu: Cho số nguyên x, trả về True nếu tồn tại số nguyên n ≥ 1 sao cho x = 1 + 2 + ... + n, ngược lại trả về False. 
Khuyến khích dùng vòng lặp đơn giản để cộng dồn tổng.

```python
def kiem_tra_tong_don_gian(x):
    tong = 0
    n = 1
    while tong < x:
        tong = tong + n
        n = n + 1
    
    return tong == x

print(kiem_tra_tong_don_gian(6))
print(kiem_tra_tong_don_gian(10))
print(kiem_tra_tong_don_gian(8))
```

Yêu cầu: Cho tọa độ ba điểm A(x1,y1), B(x2,y2), C(x3,y3), trả về True nếu chúng tạo thành một tam giác, ngược lại trả về 
False. 
Quy tắc: Chúng tạo thành tam giác nếu không thẳng hàng (diện tích có hướng khác 0). 

```python
def kiem_tra_tam_giac(x1, y1, x2, y2, x3, y3):
    dien_tich_co_huong = x1 * (y2 - y3) + x2 * (y3 - y1) + x3 * (y1 - y2)
    
    if dien_tich_co_huong != 0:
        return True
    else:
        return False

print(kiem_tra_tam_giac(0, 0, 3, 0, 0, 4))
print(kiem_tra_tam_giac(0, 0, 1, 1, 2, 2))
```

Yêu cầu: Cho các hệ số thực a, b, c, trả về tất cả các nghiệm thực của phương trình bậc hai axx + bx + c = 0 trong một danh 
sách được sắp xếp tăng dần.  
Quy tắc: - Nếu a == 0 và b != 0: trả về nghiệm duy nhất [-c / b]. - Nếu a == 0 và b == 0: trả về [] (không có nghiệm hữu hạn duy nhất nếu c == 0, hoặc vô nghiệm nếu c != 0). - Với a != 0, dùng biệt thức d = bb - 4ac: - d < 0 → [] - d == 0 → [ -b / (2a) ] - d > 0 → hai nghiệm [r1, r2] với r1 <= r2. 

```python
import math

def giai_phuong_trinh_bac_hai(a, b, c):
    if a == 0:
        if b != 0:
            return [-c / b]
        else:
            return []
    
    d = b*b - 4*a*c
    
    if d < 0:
        return []
    elif d == 0:
        return [-b / (2*a)]
    else:
        can_d = math.sqrt(d)
        r1 = (-b - can_d) / (2*a)
        r2 = (-b + can_d) / (2*a)
        
        ket_qua = [r1, r2]
        ket_qua.sort()
        return ket_qua

print(giai_phuong_trinh_bac_hai(1, -3, 2))
print(giai_phuong_trinh_bac_hai(1, 2, 1))
print(giai_phuong_trinh_bac_hai(0, 2, -4))
print(giai_phuong_trinh_bac_hai(1, 0, 1))
```

5 Yêu cầu: Cho ba số a, b, c, trả về chúng dưới dạng danh sách theo thứ tự không giảm.  


```python
def sap_xep_ba_so(a, b, c):
    danh_sach = [a, b, c]
    danh_sach.sort()
    return danh_sach

print(sap_xep_ba_so(10, 5, 8))
print(sap_xep_ba_so(1, 1, 1))
print(sap_xep_ba_so(-3, 0, -10))
```

Yêu cầu: Cho năm số a, b, c, d, e, trả về True nếu tồn tại bất kỳ ba số nào trong đó bằng nhau; ngược lại trả về False. 
Không dùng tuple hay dictionary.

```python
def kiem_tra_ba_so_bang_nhau(a, b, c, d, e):
    danh_sach = [a, b, c, d, e]
    
    for so in danh_sach:
        dem = 0
        for phan_tu in danh_sach:
            if so == phan_tu:
                dem = dem + 1
        
        if dem >= 3:
            return True
            
    return False

print(kiem_tra_ba_so_bang_nhau(1, 2, 3, 1, 1))
print(kiem_tra_ba_so_bang_nhau(1, 2, 3, 4, 5))
print(kiem_tra_ba_so_bang_nhau(7, 7, 8, 8, 7))
```

Yêu cầu: Cho ba số a, b, c, trả về giá trị ở giữa khi sắp xếp ba số theo thứ tự tăng dần (giá trị lớn thứ hai). 
Định nghĩa này vẫn đúng khi có phần tử trùng nhau (ví dụ: 2, 2, 5 → giá trị giữa là 2).

```python
def tim_so_o_giua(a, b, c):
    danh_sach = [a, b, c]
    danh_sach.sort()
    
    so_o_giua = danh_sach[1]
    
    return so_o_giua

print(tim_so_o_giua(2, 2, 5))
print(tim_so_o_giua(10, 5, 8))
print(tim_so_o_giua(1, 9, 5))
```

Yêu cầu: Cho hai ngày sinh dưới dạng (day, month, year) của người 1 và người 2, trả về 1 nếu người 1 lớn tuổi hơn (ngày 
sinh sớm hơn), 2 nếu người 2 lớn tuổi hơn, và 0 nếu bằng nhau. 
Ví dụ: 27 2 2007 so với 28 2 2007 → trả về 1 vì ngày 27 sớm hơn nên người 1 lớn tuổi hơn.

```python
def so_sanh_tuoi(d1, m1, y1, d2, m2, y2):
    nguoi_1 = [y1, m1, d1]
    nguoi_2 = [y2, m2, d2]
    
    if nguoi_1 < nguoi_2:
        return 1
    elif nguoi_1 > nguoi_2:
        return 2
    else:
        return 0

print(so_sanh_tuoi(27, 2, 2007, 28, 2, 2007))
print(so_sanh_tuoi(1, 1, 1990, 1, 1, 1990))
print(so_sanh_tuoi(10, 5, 2000, 5, 5, 2000))
```

Yêu cầu: Cho số nguyên n, trả về: - 5n + 1 nếu n mod 3 == 1 - 6n + 2 nếu n mod 3 == 2 - 7n + 3 nếu n mod 3 == 0 

```python
def tinh_gia_tri_n(n):
    so_du = n % 3
    
    if so_du == 1:
        return 5 * n + 1
    elif so_du == 2:
        return 6 * n + 2
    else:
        return 7 * n + 3

print(tinh_gia_tri_n(4))
print(tinh_gia_tri_n(5))
print(tinh_gia_tri_n(3))
```

Yêu cầu: Cho điểm số s, trả về xếp loại chữ theo quy tắc: - A+ nếu s ≥ 9 - A nếu s ≥ 8.5 - B+ nếu s ≥ 8 - B nếu s ≥ 7 - E nếu còn lại

```python
def xep_loai_diem(s):
    if s >= 9:
        return "A+"
    elif s >= 8.5:
        return "A"
    elif s >= 8:
        return "B+"
    elif s >= 7:
        return "B"
    else:
        return "E"

print(xep_loai_diem(9.5))
print(xep_loai_diem(8.2))
print(xep_loai_diem(7.5))
print(xep_loai_diem(5.0))
```

Yêu cầu: Cho một chuỗi giới tính và chiều cao tính bằng mét, trả về True nếu người đó là nam và cao hơn 1.70 m; ngược 
lại trả về False. 
Chấp nhận các cách biểu diễn nam như 'male', 'm', 'nam' (không phân biệt hoa thường).

```python
def kiem_tra_nam_cao(gioi_tinh, chieu_cao):
    danh_sach_nam = ["male", "m", "nam"]
    
    gioi_tinh_chuan = gioi_tinh.lower()
    
    if gioi_tinh_chuan in danh_sach_nam and chieu_cao > 1.70:
        return True
    else:
        return False

print(kiem_tra_nam_cao("Nam", 1.75))
print(kiem_tra_nam_cao("m", 1.65))
print(kiem_tra_nam_cao("Female", 1.80))
print(kiem_tra_nam_cao("MALE", 1.71))
```
