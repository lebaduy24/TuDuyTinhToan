Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng giữ 
nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả. 
Yêu cầu in ra đúng dạng list Python, ví dụ [1, 2, 3] 

```python
dong_nhap = input()
danh_sach_goc = [int(x) for x in dong_nhap.split()]

ket_qua = []
da_xuat_hien = set()

for so in danh_sach_goc:
    if so not in da_xuat_hien:
        ket_qua.append(so)
        da_xuat_hien.add(so)

print(ket_qua)
```

Nhập một dòng số nguyên. In ra list tổng tích lũy (phần tử i là tổng từ đầu tới vị 
trí i), giữ nguyên thứ tự

```python
dong_nhap_vao = input()
danh_sach_so = [int(x) for x in dong_nhap_vao.split()]

tong_tich_luy = []
tong_hien_tai = 0

for so in danh_sach_so:
    tong_hien_tai += so
    tong_tich_luy.append(tong_hien_tai)

print(tong_tich_luy)
```

Nhập một tuple số nguyên (một dòng) và số nguyên k (dòng kế). In ra tuple 
sau khi xoay trái k vị trí. (Giữ nguyên thứ tự tương đối sau xoay; không sắp 
xếp.)

```python
dong_dau = input()
k = int(input())

day_so = tuple(int(x) for x in dong_dau.split())
do_dai = len(day_so)

if do_dai == 0:
    print(())
else:
    k_thuc_te = k % do_dai
    ket_qua = day_so[k_thuc_te:] + day_so[:k_thuc_te]
    print(ket_qua)
```

Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo 
dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key lặp 
lại, thêm vào list của key đó 

```python
dong_nhap = input()
cac_cum_du_lieu = dong_nhap.split()

tu_dien_ket_qua = {}

for cum in cac_cum_du_lieu:
    khoa, gia_tri = cum.split(':')
    if khoa not in tu_dien_ket_qua:
        tu_dien_ket_qua[khoa] = [gia_tri]
    else:
        tu_dien_ket_qua[khoa].append(gia_tri)

print(tu_dien_ket_qua)
```

Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ..., 'zeros': 
...} (đếm số >0, <0, =0) 

```python
dong_nhap = input()
danh_sach_so = [int(x) for x in dong_nhap.split()]

duong = 0
am = 0
khong = 0

for so in danh_sach_so:
    if so > 0:
        duong += 1
    elif so < 0:
        am += 1
    else:
        khong += 1

ket_qua = {
    'positives': duong,
    'negatives': am,
    'zeros': khong
}

print(ket_qua)
```

Viết chương trình nhập vào một list các chuỗi, sau đó đếm số lần xuất hiện của từng 
chuỗi và lưu vào dictionary. 

```python
dong_nhap = input()
danh_sach_chuoi = dong_nhap.split()

tu_dien_dem = {}

for chuoi in danh_sach_chuoi:
    if chuoi in tu_dien_dem:
        tu_dien_dem[chuoi] += 1
    else:
        tu_dien_dem[chuoi] = 1

print(tu_dien_dem)
```

Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một dictionary mới 
trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ được cộng lại, nếu chỉ có 
trong một dict thì giữ nguyên. In ra dictionary mới theo thứ tự từ điển tăng dần của 
key. 

```python
chuoi_tu_dien_1 = input()
chuoi_tu_dien_2 = input()

tu_1 = eval(chuoi_tu_dien_1)
tu_2 = eval(chuoi_tu_dien_2)

ket_qua_gop = tu_1.copy()

for khoa, gia_tri in tu_2.items():
    if khoa in ket_qua_gop:
        ket_qua_gop[khoa] += gia_tri
    else:
        ket_qua_gop[khoa] = gia_tri

cac_khoa_da_sap_xep = sorted(ket_qua_gop.keys())
tu_dien_cuoi_cung = {}
for k in cac_khoa_da_sap_xep:
    tu_dien_cuoi_cung[k] = ket_qua_gop[k]

print(tu_dien_cuoi_cung)
```

Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên dương k. In ra 
list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là một tuple, hai số trong tuple 
có thể giống nhau, thứ tự xuất hiện trong danh sách của số thứ nhất nhỏ hơn thứ tự của 
số thứ hai. Các tuple trong list được sắp xếp theo thứ tự tăng dần của phần tử đầu tiên 
trong tuple.

```python
dong_dau = input()
k = int(input())

danh_sach = [int(x) for x in dong_dau.split()]
ket_qua = []
da_dung = [False] * len(danh_sach)

for i in range(len(danh_sach)):
    if da_dung[i]:
        continue
    
    for j in range(i + 1, len(danh_sach)):
        if not da_dung[j] and danh_sach[i] + danh_sach[j] == k:
            ket_qua.append((danh_sach[i], danh_sach[j]))
            da_dung[i] = True
            da_dung[j] = True
            break

ket_qua.sort()
print(ket_qua)
```

[Tách số chẵn và lẻ từ tuple] 
Nhập vào một tuple chứa các số nguyên. Hãy tạo ra hai tuple mới: một chứa 
tất cả các số chẵn và một chứa tất cả các số lẻ từ tuple ban đầu, giữ nguyên 
thứ tự tương đối. In ra hai tuple này trên hai dòng riêng biệt.

```python
dong_nhap = input()
day_so = [int(x) for x in dong_nhap.split()]

chan = []
le = []

for so in day_so:
    if so % 2 == 0:
        chan.append(so)
    else:
        le.append(so)

print(tuple(chan))
print(tuple(le))
```

[Tìm phần tử xuất hiện nhiều nhất] 
Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong list. 
Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ nhất 
trong số đó. 

```python
dong_nhap = input()
danh_sach = [int(x) for x in dong_nhap.split()]

tu_dien_dem = {}
for so in danh_sach:
    tu_dien_dem[so] = tu_dien_dem.get(so, 0) + 1

so_lan_nhieu_nhat = -1
ket_qua = None

for so in tu_dien_dem:
    luot_dem = tu_dien_dem[so]
    if luot_dem > so_lan_nhieu_nhat:
        so_lan_nhieu_nhat = luot_dem
        ket_qua = so
    elif luot_dem == so_lan_nhieu_nhat:
        if so < ket_qua:
            ket_qua = so

print(ket_qua)
```

[Đảo ngược key và value của dictionary] 
Nhập vào một dictionary có value là duy nhất (không trùng nhau). Hãy tạo một 
dictionary mới bằng cách đảo ngược key và value của dictionary ban đầu. In 
ra dictionary mới. 

```python
dong_nhap = input().split()
tu_dien_moi = {}

for i in range(0, len(dong_nhap), 2):
    khoa = dong_nhap[i]
    gia_tri = dong_nhap[i+1]
    tu_dien_moi[gia_tri] = khoa

print(tu_dien_moi)
```

[Giao của hai danh sách] 
Nhập vào hai dòng, mỗi dòng là một danh sách các số nguyên. Hãy tìm các 
phần tử chung (giao) của hai danh sách này và in ra một list mới chứa các 
phần tử chung đó, không trùng lặp.

```python
dong_dau = input()
n = int(input())

danh_sach_tu = dong_dau.split()
ket_qua = []

for tu in danh_sach_tu:
    if len(tu) > n:
        ket_qua.append(tu)

print(ket_qua)
```

[Lọc dictionary theo giá trị] 
Nhập vào một dictionary có key là chuỗi và value là số nguyên, và một số 
nguyên k ở dòng tiếp theo. Hãy tạo một dictionary mới chỉ chứa các cặp key
value từ dictionary ban đầu mà có value lớn hơn k. In ra dictionary mới.

```python
chuoi_danh_sach = input()
khoa_k = input()

danh_sach_tu_dien = eval(chuoi_danh_sach)

tu_dien_lon_nhat = danh_sach_tu_dien[0]

for tu_dien_hien_tai in danh_sach_tu_dien:
    if tu_dien_hien_tai[khoa_k] > tu_dien_lon_nhat[khoa_k]:
        tu_dien_lon_nhat = tu_dien_hien_tai

print(tu_dien_lon_nhat)
```
