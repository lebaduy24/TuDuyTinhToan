Bài 1: Nhập vào 2 cạnh của một hình chữ nhật. In ra màn hình diện tích và chu

Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là 
độ dài ba cạnh hay không, nếu  không thì in  ra màn  hình  'Khong la ba  canh 
cua mot tam  giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, 
Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn

```python
import math
a,b,c = map(float,input().split())
if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        print ("Tam giác đều")
    elif a == b or a == c or b == c:
        print ("Tam giác cân")
    elif (a**2 + b**2 == c**2) or (b**2 + c**2 == a**2) or (a**2 + c**2 == b**2):
        if a == b or b == c or a == c:
              print ("Tam giác vuông cân")
        else:
              print ("Tam giác vuông")
    else:
        print ("Tam giác thường")
    p = (a+b+c)/2
    dien_tich = math.sqrt(p*(p-a)*(p-b)*(p-c))
    chu_vi = a+b+c
    print ("Diện tích tam giác là:", dien_tich)
    print ("Chu vi tam giác là:", chu_vi)
else:
     print ("Không phải tam giác")
    
        
```

Bài 8: Giải hệ phương trình tuyến tính: 
{ax + by = m 
; cx + dy = n 

```python
a,b,c,d = map(float,input("Nhập hệ số tương ứng:").split())
m,n = map(float,input("Nhập hệ số tự do:").split())
if a*d != b*c: # a/c khác b/d
    x = (m*d - b*n) / (a*d -b*c)
    y = (a*n - c*m) / (a*d -b*c)
    print (f"Hệ phương trình có nghiệm x={x} và y={y}")
else:
    if (a*d == b*c) and (b*n==m*d) and (a*n==m*c):
        print ("Phương trình vô số nghiệm")
    else:
        print ("Phương trinh vô nghiệm")

```

Bài 9: Nhập vào thời gian 1 công việc nào đó là x giây. Hãy chuyển đổi và viết 
ra màn hình số thời gian trên dưới dạng bằng bao nhiêu giờ, bao nhiêu phút, bao 
nhiêu giây. 

```python
thoi_gian = float(input())
gio = thoi_gian // 3600
giay_con_lai = thoi_gian % 3600
phut = giay_con_lai // 60
giay = giay_con_lai % 60
print (f"Thời gian 1 công việc là {gio}h {phut}p {giay}s ")
```

Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào một 
điểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
a,b = map(float,input("Nhập tâm đường tròn").split())
bk = float(input("Nhập bán kính đường tròn"))
m,n = map(float,input("Nhập một điểm để kiểm tra có thuộc đường tròn hay không").split())
if (m-a)**2 + (n-b)**2 == bk**2:
    print ("Điểm này thuộc đường tròn")
else:
    print ("Điểm này không thuộc đường tròn")
```
